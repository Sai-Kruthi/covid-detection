## Checking All Files

```python
import cv2
def loadModel():
    from tensorflow.keras.models import load_model
    model = load_model("model\M76.h5")
    return model

def predict(image):
    model = loadModel()
    img = cv2.resize(image, (180,180))
    import numpy as np
    img1 = np.reshape(img,(1,180,180,3))
    result = model.predict(img1)
    total = np.sum(result)
   
    res = []
    for i in result[0]:
        res.append(i/total)
    return res
    
```

```python
import os
```

```python
def forFolder(folder1,folder2):
    folder = folder1 + folder2
    count = 0
    l = []
    s = 0
    temp = 0
    for foldername in os.listdir(folder):
        
        loc = os.path.join(folder , foldername)
        
        for filename in os.listdir(loc):
            img = cv2.imread(os.path.join(loc, filename))
            
            result = predict(img)
            s1 = result.index(max(result))
            l1 = [s,s1,result[0]*100 , result[1]*100 , s == s1]
            l.append(l1)
            print(temp)
            temp += 1
            if s == s1:
                count += 1
            ## print(foldername, result[0]*100,"% ",result[1]*100,"% " )
        s += 1
    print(l)
    return l,count
```

```python
### 
# X holds the images read as input
# Y holds the labels of the images

dataset_path = "Dataset/"
sub_path = "test/"
ans,c = forFolder(dataset_path,sub_path)
```
